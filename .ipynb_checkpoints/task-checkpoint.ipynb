```python
import numpy as np
import pandas as pd
```

```python
PB = pd.read_csv("purchase_behaviour.csv")
TD = pd.read_csv("transaction_data.csv")
```

```python
PB
```

```python
TD
```

```python
df = pd.merge(TD, PB, on="LYLTY_CARD_NBR", how="inner")
```

```python
df
```

```python
customer_sales = df.groupby("LYLTY_CARD_NBR")["TOT_SALES"].sum().reset_index()
customer_sales
```

```python
customer_profile = pd.merge(customer_sales, PB, on="LYLTY_CARD_NBR", how="inner")
customer_profile
```

```python
top_customers = customer_profile.nlargest(int(0.1 * len(customer_profile)), "TOT_SALES")
top_customers
```

```python
summary = top_customers.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["TOT_SALES"].sum().reset_index()
summary
```

```python
top_products = df.groupby(["PROD_NBR", "PROD_NAME"])["TOT_SALES"].sum().reset_index()
```

```python
top3_products = top_products.nlargest(3, "TOT_SALES") 
top3_products
```

```python
loyal_customers = df.groupby("LYLTY_CARD_NBR")["TXN_ID"].count().reset_index()
loyal_customers = loyal_customers.nlargest(int(0.1 * len(loyal_customers)), "TXN_ID")
loyal_customers
```

```python
loyal_customer_profiles = pd.merge(loyal_customers, PB, on="LYLTY_CARD_NBR", how="inner")
loyal_customer_profiles
```

```python
loyal_summary = loyal_customer_profiles.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["TXN_ID"].count().reset_index()
loyal_summary
```
